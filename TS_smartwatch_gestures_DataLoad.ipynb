```python
import os
def list_files(filepath, filetype):
   paths = []
   for root, dirs, files in os.walk(filepath):
      for file in files:
         if file.lower().endswith(filetype.lower()):
            paths.append(os.path.join(root, file))
   return(paths)
```

```python
data_files = list_files('/content/drive/MyDrive/data/smart_watch/smartwatch_gestures','txt')
```

```python
# Create new dataframe to hold data
# Create columns for user and gesture from dir tree: /<user>/<gesture>/XX.txt
# File column labels: System.currentTimeMillis() System.nanoTime() event.timestamp event.values[0] event.values[1] event.values[2]
df = pd.DataFrame(columns = ['user', 'gesture', 'trial', 'step', 'sys_current_time', 'sys_nanotime', 'event_timestamp','X', 'Y', 'Z'])

df
```

```python
FILEPATH = '/content/drive/MyDrive/data/smart_watch/smartwatch_gestures'
FILETYPE = 'txt'


# identify all files with data
paths = []
for root, dirs, files in os.walk(FILEPATH):
  for file in files:
    if file.lower().endswith(FILETYPE.lower()):
      paths.append(os.path.join(root, file))



i = len(FILEPATH.split('/')) 

for x in paths:
  user = x.split('/')[i]
  gesture = x.split('/')[i+1]
  
  # print(x)

  row_elements = np.array([])
  row_elements = np.loadtxt(x)

  temp_df = pd.DataFrame(row_elements, columns = ['sys_current_time', 'sys_nanotime', 'event_timestamp','X', 'Y', 'Z'])

  user_info = [user for y in range(len(row_elements))]
  gesture_info = [gesture for y in range(len(row_elements))]

  temp_df['user'] = user_info
  temp_df['gesture'] = gesture_info

  temp_df = temp_df[['user', 'gesture', 'sys_current_time', 'sys_nanotime', 'event_timestamp','X', 'Y', 'Z']]
  df = pd.concat([df, temp_df])




```

```python
df.to_csv('smartwatch_gestures.csv')
```
